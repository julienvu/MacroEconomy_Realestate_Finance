```python
def eval_object(v):
    operations = { '+': v['a']+v['b'],
                   '-': v['a']-v['b'],
                   '/': v['a']/v['b'],
                   '*': v['a']*v['b'],
                   '%': v['a']%v['b'],
                   '**': v['a']**v['b']}
    if operations.keys()=='+':
        return operations['+']
    if operations.keys()=='-':
        return operations['-']
    if operations.keys()=='/':
        return operations['/']
    if operations.keys()=='*':
        return operations['*']
    if operations.keys()=='%':
        return operations['%']
    if operations.keys()=='**':
        return operations['**']
```

Write a function taking a list of lists as an argument and returning the number of unique arrays that can be formed by picking exactly one element from each subarray (the first element from the first array the second from the second etc).

**Example**

`solve([[1,2],[4],[5,6]]) = 4`,
because it results in only 4 possiblites, which are `[1,4,5],[1,4,6],[2,4,5],[2,4,6]`.

`solve([[1,2],[4,4],[5,6,6]]) = 4`, 
since the extra outcomes are just duplicates.

You can use NumPy to solve this problem.

```python

def solve(arrays):
    N = len(arrays)
    compt=0
    for i in range(N):
        combi = []
        for j in range(N):
            if (i >> j) % 2 == 1:
                combi.append(arrays[j])
                compt+=1
        
    return compt*2
    
```

```python
print(solve([[1,2],[4,4],[5,6,6]]))
print(solve([[1,2],[4,4],[5,6,7]])) # correct answer is 6
print(solve([[1,2],[4,6]])) # correct answer is 4
grade += 0
```
